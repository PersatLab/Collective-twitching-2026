# Nematic correlation function


```python
# Nematic correlation fucntion as a function of the distance between cell pairs for the first frame
number_of_distance_points = 40
analysis_frame = 0
semilog_plot = False
save = True
save_to_csv = True
outputDirectory = folder_manager.segmentsanalysisdir

for name_movie in movies_masks_edited:
    largest_distance = min(np.shape(movies_masks_edited[name_movie][0]))

    distances = np.logspace(1, np.log10(largest_distance), num=number_of_distance_points)

    mask_video = movies_masks_edited[name_movie]
    first_frame = mask_video[analysis_frame]
    
    global_result = sau.plot_nematic_correlation_function(
        first_frame, 
        distances, 
        pixel_scale, 
        outputDirectory, 
        name_movie, 
        analysis_frame,  # Pass the frame index here
        plot = True,
        semilog=semilog_plot, 
        save=save
    )
```

```python
### Fit of the nematic correlation function

import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.optimize import curve_fit

outputDirectory = folder_manager.segmentsanalysisdir
graphe_outputDirectory = folder_manager.graphdir
combined = True

bacterial_width = 0.8 #µm

fit_results = pd.DataFrame(columns=["Filename", "Fitting Equation", "a", "b", "c"])

for name_movie in movies_masks_edited:
    csv_path = os.path.join(outputDirectory, name_movie, 'nematic_correlation')
    df = pd.read_csv(csv_path + '/' + name_movie + '_nematic_correlation_metadata_frame_0.csv', sep=',')    
    df = df.dropna()
    df = df.sort_values(by="radius")

    if combined == False:
        # Prepare a DataFrame to store results
        fit_results = pd.DataFrame(columns=["Filename", "Fitting Equation", "a", "b", "c"])
    
    # Step 1. Plot raw curve
    # Compute the average for each radius (to smooth individual curves)
    #df = df.groupby("radius", as_index=False)["local_order_parameter"].mean()

    plt.figure(figsize=(4,4))
    plt.plot(
        df["radius"] / bacterial_width,
        df["mean_nematic_correlation"],
        alpha=1,
        color = 'black',
        linewidth=5,
        label='Single experimental curve'
    )

    #plt.title(f"Strain {strain} - {sc}")
    plt.xlim(0, 35)
    plt.ylim(0, 1)
    plt.xlabel("Distance", fontsize=24, fontweight='bold', fontname='Arial')
    plt.ylabel("Nematic Correlation", fontsize=24, fontweight='bold', fontname='Arial')
    plt.xticks(ticks=[0, 10, 20, 30])
    plt.yticks(ticks=[0, 0.25, 0.5, 0.75, 1])
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(1.25)
    plt.gca().spines['bottom'].set_linewidth(1.25)
    plt.gca().set_box_aspect(1)
    plt.tick_params(axis='x', labelsize=18)
    plt.tick_params(axis='y', labelsize=18)
    plt.tight_layout()

    # Step 2. Fit only for radius > 2 µm
    df_fit = df[df["radius"] > 2] 

    X = np.array(df_fit["radius"] / bacterial_width)
    Y = np.array(df_fit["mean_nematic_correlation"])

    # Fit using the extended model with offset
    popt, pcov = curve_fit(sau.model_exp, X, Y, p0=[1, 10, 0])

    # Generate the fitted curve
    Y_fit = sau.model_exp(X, *popt)

    # Extract parameters
    a, b, c = popt

    # Plot the fitted curve
    plt.plot(df["radius"] / bacterial_width, sau.model_exp(np.array(df["radius"] / bacterial_width), *popt), linewidth = 3, linestyle='--', color='red', 
             label='Fit: a=' + str(round(popt[0], 2)) + 
             ', b=' + str(round(popt[1], 2)) + 
             ', c=' + str(round(popt[2], 2)))
  
    save_path = os.path.join(folder_manager.graphdir, name_movie, 'mean_nematic_correlation')
    save_fig_path = os.path.join(graphe_outputDirectory, name_movie, "nematic_correlation")
    os.makedirs(save_fig_path, exist_ok=True)
    plt.savefig(os.path.join(save_fig_path, f"{name_movie}_fit.png"), dpi=300, bbox_inches='tight', transparent=True)
    plt.show()
    plt.close()

    # Extract parameters
    a = popt[0]
    b = popt[1]
    c = popt[2]

    # Add results to the DataFrame
    fit_results = pd.concat([fit_results, pd.DataFrame({
        "Filename": [name_movie],
        "Fitting Equation": ["a * exp(-Radius/b) + c"],
        "a": [round(a, 4)],
        "b": [round(b, 4)],
        "c": [round(c, 4)]
    })], ignore_index=True)

    if combined == False:
        # Save results to CSV
        output_csv_path = os.path.join(csv_path, name_movie + "_fitting_results.csv")
        fit_results.to_csv(output_csv_path, index=False)
        print(f"Fitting results saved to {output_csv_path}")

if combined == True:
    # Save results to CSV
    output_csv_path = os.path.join(outputDirectory, "fitting_results_combined.csv")
    fit_results.to_csv(output_csv_path, index=False)
    print(f"Fitting results saved to {output_csv_path}")
```
