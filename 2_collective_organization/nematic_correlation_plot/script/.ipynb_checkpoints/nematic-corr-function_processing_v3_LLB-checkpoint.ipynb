```python
def replace_legend_label(label):
    strain_labels = {
        '177': 'WT',
        '232': r'$\Delta\mathit{pilH}$',
        '459': r'$\Delta\mathit{pilG}$',
        '1047': r'$\Delta\mathit{pilA}$',
        '1695': r'$\Delta\mathit{pilA\ cpdA}$',
        '337' : r'$\Delta\mathit{cpdA}$',
        '436' : r'$\Delta\mathit{pilH\ cyaB}$'
    }
    return strain_labels.get(label, label)  # Return the mapped label, or the original if not found
```

```python
## For figures - draft (240306)

import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.lines import Line2D

# Folder paths
data_folder = "dataset_NCF_high-density"  # Folder where the CSV files are located
figures_folder = "figures"  # Folder where figures will be saved
#theory_folder = "/Users/laureleblanc/Desktop/PersatLAB/4_Analysis/Code/6_Order_parameter/dataset-simulation/"
theory_folder = "/Users/laureleblanc/Desktop/PersatLAB/5_Analysis/Code/6_Order_parameter/dataset-simulation-v2-250729-high/"

strain_labels = {
        '232': r'$P_{\mathrm{CR}} = 0$', #r'$\Delta\mathit{pilH}$',
        '177': r'$P_{\mathrm{CR}} = 0.2$',
        '459': r'$P_{\mathrm{CR}} = 0.8$',
        '104': r'$\Delta\mathit{pilA}$',
        '169': r'$\Delta\mathit{pilA\ cpdA}$',
    }

# Define colors for strains
strain_colors = {
    "177": "magenta",
    "232": "deepskyblue",
    "459": "darkorange",
    "1047": "lightgray",
    "1695": "gray",
}

# Surface coverage options
sc_options = ["SC10pc", "SC30pc", "SC70pc"]

# Choose which strains to plot for surface coverage graphs
strains_to_plot = ["177", "232"] #, "459", "1047", "1695"]  # Example of selected strains

# Option for plotting individual curves: "gray" or "colors"
plot_single_curves_in_gray = True  # Change to True to plot single curves in gray

# Step 1: Load the data and group by strain and SC
data_by_strain_sc = {}  # Dictionary to hold data grouped by strain and SC
files = [f for f in os.listdir(data_folder) if f.endswith(".csv")]

for file in files:
    filepath = os.path.join(data_folder, file)
    try:
        df = pd.read_csv(filepath)
        df = df.dropna()
        df = df.sort_values(by="radius")
        #print(df.head(5))

        # Normalize radius by bacterial width 0.8
        df["radius"] = df["radius"] / 0.8

        # Compute the average for each radius (to smooth individual curves)
        df = df.groupby("radius", as_index=False)["nematic_correlation"].mean()

        # Extract strain and SC from filename
        strain_found = None
        sc_found = None #"SC10pc"
        
        # Extract strain (appears after the date in the format yymmdd_strain)
        for strain in ["177", "232"]: #, "459", "1047", "1695"]:
            if strain in file.split('_')[1]:  # Check if the strain is after the date
                strain_found = strain
                break
        
        # Extract surface coverage (e.g., _SC10pc)
        for sc in sc_options:
            if sc in file:
                sc_found = sc
                print(sc)
                break
        
        # Ensure both strain and SC are found before adding the data
        if strain_found and sc_found:
            key = (strain_found, sc_found)
            if key not in data_by_strain_sc:
                data_by_strain_sc[key] = []
            data_by_strain_sc[key].append((df, file))  # Store the dataframe and filename
        else:
            print(f"Skipping file {file} as it doesn't match a known strain or SC.")
    
    except Exception as e:
        print(f"Error loading file {file}: {e}")


### Load theory data
# Dictionary to store theoretical curves
theory_data_by_strain_sc = {}

# Load theoretical data and map to (strain, SC)
theory_files = [f for f in os.listdir(theory_folder) if f.endswith(".csv")]

for file in theory_files:
    filepath = os.path.join(theory_folder, file)
    try:
        df = pd.read_csv(filepath, header=None)
        df.columns = ["radius", "nematic_correlation"]
        df = df.dropna()
        df = df.sort_values(by="radius")


        # Detect strain and SC from filename
        strain_found = None
        sc_found = None

        for strain in strain_colors.keys():
            if strain in file:
                strain_found = strain
                break
        for sc in sc_options:
            if sc in file:
                sc_found = sc
                break

        if strain_found and sc_found:
            theory_data_by_strain_sc[(strain_found, sc_found)] = df
        else:
            print(f"Skipping theory file {file} — strain or SC not found.")

    except Exception as e:
        print(f"Error loading theory file {file}: {e}")


# Step 2: Plot individual curves and averages for each strain and SC
for (strain, sc), dfs in data_by_strain_sc.items():
    
    plt.figure(figsize=(4, 4))
    
    # Define a colormap to assign different colors to individual curves
    colormap = cm.get_cmap("tab20", len(dfs))  # Use "tab20" for distinct colors

    # Plot individual curves with unique colors
    for i, (df, file) in enumerate(dfs):
        if plot_single_curves_in_gray:
            color = "gray"  # Gray for all individual curves
        else:
            color = colormap(i)  # Assign a unique color from the colormap
        
        plt.plot(
            df["radius"],
            df["nematic_correlation"],
            color=color,
            alpha=0.7,
            linewidth=1.5,
            label=file[:10] + '_' + sc + '_' + file[42:58]  # Use the full filename as the legend label for individual curves
        )

    # Calculate and plot the average curve for this strain and SC
    combined_data = pd.concat([df for df, _ in dfs])
    avg_curve = combined_data.groupby("radius")["nematic_correlation"].mean()
    plt.plot(
        avg_curve.index,
        avg_curve.values,
        color=strain_colors[strain],
        linewidth=4,
        label=f"Average - Strain {strain} - {sc}"
    )

    # Plot theory curve if available
    theory_key = (strain, sc)
    if theory_key in theory_data_by_strain_sc:
        theory_df = theory_data_by_strain_sc[theory_key]
        plt.plot(
            theory_df["radius"],
            theory_df["nematic_correlation"],
            linestyle='--',
            linewidth=4,
            color=strain_colors[strain],
            label=f"Theory - Strain {strain} - {sc}"
        )


    # Add title, labels, and legend
    plt.title(f"Strain {strain} - {sc}")
    plt.ylim(0, 1)
    plt.xlim(0, 35)
    plt.xlabel("") #"Distance")
    plt.ylabel("") #"Nematic Correlation")
    plt.legend(frameon=False)  # Ensure all curves appear in the legend
    plt.tight_layout()
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(1.25)
    plt.gca().spines['bottom'].set_linewidth(1.25)
    plt.savefig(os.path.join(figures_folder, f"strain_{strain}_{sc}.png"))
    plt.savefig(os.path.join(figures_folder, f"strain_{strain}_{sc}.svg"))
    plt.close()

# Step 3: Plot average curves for each strain, across all SCs (transparency based on SC)
for strain in ["177", "232", "459", "1047", "1695"]:  # Iterate over all strains
    #plt.figure(figsize=(5.5, 5))
    plt.figure(figsize=(4.5, 5))
    
    for sc in sc_options:
        key = (strain, sc)
        
        # Only plot if there is data for the strain and SC
        if key in data_by_strain_sc:
            dfs = data_by_strain_sc[key]
            combined_data = pd.concat([df for df, _ in dfs])
            avg_curve = combined_data.groupby("radius")["nematic_correlation"].mean()
            
            # Define transparency based on surface coverage
            alpha_value = 1.0 if sc == "SC70pc" else 0.6 if sc == "SC30pc" else 0.5  # Adjust transparency
            
            label_text = "Low density" if sc == "SC10pc" else "High density" if sc == "SC70pc" else "30% (Average)"
            plt.plot(
                avg_curve.index,
                avg_curve.values,
                color=strain_colors[strain],
                linewidth=4,
                alpha=alpha_value,
                label=label_text
            )

            # Plot theory curve if available
            theory_key = (strain, sc)
            if theory_key in theory_data_by_strain_sc:
                theory_df = theory_data_by_strain_sc[theory_key]
                plt.plot(
                    theory_df["radius"],
                    theory_df["nematic_correlation"],
                    linestyle='--',
                    linewidth=4,
                    color=strain_colors[strain],
                    alpha=alpha_value,
                    label=f"Theory - {sc}"
                )

    #plt.title(f"Average Curves for Strain {strain} at Different SCs")
    plt.ylim(0, 1)
    plt.xlim(0, 35)
    plt.xlabel("Distance", fontsize=22, fontweight='bold', fontname='Arial')
    plt.ylabel("Nematic Correlation", fontsize=22, fontweight='bold', fontname='Arial')
    plt.xticks(ticks=[0, 10, 20, 30], fontsize=18)
    plt.yticks(ticks=[0.0, 0.25, 0.5, 0.75, 1], fontsize=18)
    plt.legend(loc='upper right', fontsize=18, frameon=False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(1.25)
    plt.gca().spines['bottom'].set_linewidth(1.25)
    plt.tight_layout()
    plt.savefig(os.path.join(figures_folder, f"average_strain_{strain}_all_SCs_overlay-with-simus.png"), dpi=300, transparent=True)
    plt.savefig(os.path.join(figures_folder, f"average_strain_{strain}_all_SCs_overlay-with-simus.svg"), dpi=300, transparent=True)
    plt.close()

# Define plotting order (e.g., by strain ID)
#plot_order = ["1047", "1695", "459", "232", "177"]
#plot_order = ["459", "232", "177"]
plot_order = ["232", "177"]

# Define legend order (custom order for display)
#legend_order = ["232", "177", "459", "1047", "1695"]
#legend_order = ["232", "177", "459"]
legend_order = ["232", "177"]

# Step 4: Plot average curves for each surface coverage, across all strains (based on selected strains)
for sc in sc_options:
    plt.figure(figsize=(4, 4))

    # Plot strains in the desired plot order
    for strain in plot_order:
        if strain not in strains_to_plot:
            continue
        key = (strain, sc)
        if key in data_by_strain_sc:
            dfs = data_by_strain_sc[key]
            combined_data = pd.concat([df for df, _ in dfs])
            avg_curve = combined_data.groupby("radius")["nematic_correlation"].mean()
            #plt.plot(
                #avg_curve.index,
                #avg_curve.values,
                #color=strain_colors[strain],
                #linewidth=4,
                #label=strain  # Temporary label
            #)

            # Plot theory curve if available
            theory_key = (strain, sc)
            if theory_key in theory_data_by_strain_sc:
                theory_df = theory_data_by_strain_sc[theory_key]
                plt.plot(
                    theory_df["radius"],
                    theory_df["nematic_correlation"],
                    linestyle='--',
                    linewidth=4,
                    color=strain_colors[strain],
                    label=f"Theory - Strain {strain}"
                )


    legend_elements = []
    for strain in legend_order:
        if strain in strains_to_plot:
            legend_elements.append(
                Line2D(
                    [0], [0],
                    color=strain_colors[strain],
                    lw=4, #strain_linewidths.get(strain, 6),  # or a fixed linewidth
                    linestyle='--',  # force plain line in legend
                    label=strain_labels[strain] #replace_legend_label(strain)
                )
            )

    plt.legend(handles=legend_elements, loc='upper right', fontsize=18, frameon=False, bbox_to_anchor=(1.05, 1.05))




    # Plot formatting
    plt.xlim(0, 35)
    plt.ylim(0, 1)
    plt.xlabel('') #"Distance", fontsize=22, fontweight='bold', fontname='Arial')
    plt.ylabel('') #"Nematic Correlation", fontsize=22, fontweight='bold', fontname='Arial')
    plt.xticks(ticks=[0, 10, 20, 30], fontsize=18)
    plt.yticks(ticks=[0.0, 0.25, 0.5, 0.75, 1], fontsize=18)
    plt.tick_params(axis='x', labelsize=18)
    plt.tick_params(axis='y', labelsize=18)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(1.25)
    plt.gca().spines['bottom'].set_linewidth(1.25)
    plt.gca().set_box_aspect(1)
    plt.tight_layout()
    plt.savefig(os.path.join(figures_folder, f"average_all_strains_{sc}_overlay-with-simus.png"), transparent=True, dpi=300)
    plt.savefig(os.path.join(figures_folder, f"average_all_strains_{sc}_overlay-with-simus.svg"), transparent=True)
    #plt.close()

```

```python
## For figures - revisions draft

import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.lines import Line2D

# Folder paths
data_folder = "/Users/laureleblanc/Desktop/PersatLAB/5_Analysis/Code/6_Order_parameter/dataset_NCF_revisions/"  # Folder where the CSV files are located
figures_folder = "figures"  # Folder where figures will be saved
#theory_folder = "/Users/laureleblanc/Desktop/PersatLAB/4_Analysis/Code/6_Order_parameter/dataset-simulation/"
theory_folder = "/Users/laureleblanc/Desktop/PersatLAB/5_Analysis/Code/6_Order_parameter/data-simulations-revisions/"

strain_labels = {
        '232': r'$P_{\mathrm{CR}} = 0$', #r'$\Delta\mathit{pilH}$',
        '177': r'$P_{\mathrm{CR}} = 0.2$',
        '459': r'$P_{\mathrm{CR}} = 0.8$',
        '104': r'$\Delta\mathit{pilA}$',
        '169': r'$\Delta\mathit{pilA\ cpdA}$',
        '337': r'$P_{\mathrm{CR}}=0.2$, $AR=4.95$',
        '436': r'$P_{\mathrm{CR}}=0$, $AR=3.45$'
    
    
    }

# Define colors for strains
strain_colors = {
    "177": "magenta",
    "232": "deepskyblue",
    "459": "darkorange",
    "1047": "lightgray",
    "1695": "gray",
    "337": "darksalmon",
    "436":"mediumblue"
}

# Surface coverage options
sc_options = ["SC10pc", "SC30pc", "SC70pc"]

# Choose which strains to plot for surface coverage graphs
strains_to_plot = ["337", "436"] #, "459", "1047", "1695"]  # Example of selected strains

# Option for plotting individual curves: "gray" or "colors"
plot_single_curves_in_gray = True  # Change to True to plot single curves in gray

# Step 1: Load the data and group by strain and SC
data_by_strain_sc = {}  # Dictionary to hold data grouped by strain and SC
files = [f for f in os.listdir(data_folder) if f.endswith(".csv")]

for file in files:
    filepath = os.path.join(data_folder, file)
    try:
        df = pd.read_csv(filepath)
        df = df.dropna()
        df = df.sort_values(by="radius")
        #print(df.head(5))

        # Normalize radius by bacterial width 0.8
        df["radius"] = df["radius"] / 0.8

        # Compute the average for each radius (to smooth individual curves)
        df = df.groupby("radius", as_index=False)["nematic_correlation"].mean()

        # Extract strain and SC from filename
        strain_found = None
        sc_found = None #"SC10pc"
        
        # Extract strain (appears after the date in the format yymmdd_strain)
        for strain in ["337", "436"]: #, "459", "1047", "1695"]:
            if strain in file.split('_')[1]:  # Check if the strain is after the date
                strain_found = strain
                break
        
        # Extract surface coverage (e.g., _SC10pc)
        for sc in sc_options:
            if sc in file:
                sc_found = sc
                print(sc)
                break
        
        # Ensure both strain and SC are found before adding the data
        if strain_found and sc_found:
            key = (strain_found, sc_found)
            if key not in data_by_strain_sc:
                data_by_strain_sc[key] = []
            data_by_strain_sc[key].append((df, file))  # Store the dataframe and filename
        else:
            print(f"Skipping file {file} as it doesn't match a known strain or SC.")
    
    except Exception as e:
        print(f"Error loading file {file}: {e}")


### Load theory data
# Dictionary to store theoretical curves
theory_data_by_strain_sc = {}

# Load theoretical data and map to (strain, SC)
theory_files = [f for f in os.listdir(theory_folder) if f.endswith(".csv")]

for file in theory_files:
    filepath = os.path.join(theory_folder, file)
    try:
        df = pd.read_csv(filepath, header=None)
        df.columns = ["radius", "nematic_correlation"]
        df = df.dropna()
        df = df.sort_values(by="radius")


        # Detect strain and SC from filename
        strain_found = None
        sc_found = None

        for strain in strain_colors.keys():
            if strain in file:
                strain_found = strain
                break
        for sc in sc_options:
            if sc in file:
                sc_found = sc
                break

        if strain_found and sc_found:
            theory_data_by_strain_sc[(strain_found, sc_found)] = df
        else:
            print(f"Skipping theory file {file} — strain or SC not found.")

    except Exception as e:
        print(f"Error loading theory file {file}: {e}")


# Step 2: Plot individual curves and averages for each strain and SC
for (strain, sc), dfs in data_by_strain_sc.items():
    
    plt.figure(figsize=(4, 4))
    
    # Define a colormap to assign different colors to individual curves
    colormap = cm.get_cmap("tab20", len(dfs))  # Use "tab20" for distinct colors

    # Plot individual curves with unique colors
    for i, (df, file) in enumerate(dfs):
        if plot_single_curves_in_gray:
            color = "gray"  # Gray for all individual curves
        else:
            color = colormap(i)  # Assign a unique color from the colormap
        
        plt.plot(
            df["radius"],
            df["nematic_correlation"],
            color=color,
            alpha=0.7,
            linewidth=1.5,
            label=file[:10] + '_' + sc + '_' + file[42:58]  # Use the full filename as the legend label for individual curves
        )

    # Calculate and plot the average curve for this strain and SC
    combined_data = pd.concat([df for df, _ in dfs])
    avg_curve = combined_data.groupby("radius")["nematic_correlation"].mean()
    plt.plot(
        avg_curve.index,
        avg_curve.values,
        color=strain_colors[strain],
        linewidth=4,
        label=f"Average - Strain {strain} - {sc}"
    )

    # Plot theory curve if available
    theory_key = (strain, sc)
    if theory_key in theory_data_by_strain_sc:
        theory_df = theory_data_by_strain_sc[theory_key]
        plt.plot(
            theory_df["radius"],
            theory_df["nematic_correlation"],
            linestyle='--',
            linewidth=4,
            color=strain_colors[strain],
            label=f"Theory - Strain {strain} - {sc}"
        )


    # Add title, labels, and legend
    plt.title(f"Strain {strain} - {sc}")
    plt.ylim(0, 1)
    plt.xlim(0, 35)
    plt.xlabel("") #"Distance")
    plt.ylabel("") #"Nematic Correlation")
    plt.legend(frameon=False)  # Ensure all curves appear in the legend
    plt.tight_layout()
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(1.25)
    plt.gca().spines['bottom'].set_linewidth(1.25)
    plt.savefig(os.path.join(figures_folder, f"strain_{strain}_{sc}.png"))
    plt.savefig(os.path.join(figures_folder, f"strain_{strain}_{sc}.svg"))
    plt.close()

# Step 3: Plot average curves for each strain, across all SCs (transparency based on SC)
for strain in ["177", "232", "459", "1047", "1695", "337", "436"]:  # Iterate over all strains
    #plt.figure(figsize=(5.5, 5))
    plt.figure(figsize=(4.5, 5))
    
    for sc in sc_options:
        key = (strain, sc)
        
        # Only plot if there is data for the strain and SC
        if key in data_by_strain_sc:
            dfs = data_by_strain_sc[key]
            combined_data = pd.concat([df for df, _ in dfs])
            avg_curve = combined_data.groupby("radius")["nematic_correlation"].mean()
            
            # Define transparency based on surface coverage
            alpha_value = 1.0 if sc == "SC70pc" else 0.6 if sc == "SC30pc" else 0.5  # Adjust transparency
            
            label_text = "Low density" if sc == "SC10pc" else "High density" if sc == "SC70pc" else "30% (Average)"
            plt.plot(
                avg_curve.index,
                avg_curve.values,
                color=strain_colors[strain],
                linewidth=4,
                alpha=alpha_value,
                label=label_text
            )

            # Plot theory curve if available
            theory_key = (strain, sc)
            if theory_key in theory_data_by_strain_sc:
                theory_df = theory_data_by_strain_sc[theory_key]
                plt.plot(
                    theory_df["radius"],
                    theory_df["nematic_correlation"],
                    linestyle='--',
                    linewidth=4,
                    color=strain_colors[strain],
                    alpha=alpha_value,
                    label=f"Theory - {sc}"
                )

    #plt.title(f"Average Curves for Strain {strain} at Different SCs")
    plt.ylim(0, 1)
    plt.xlim(0, 35)
    plt.xlabel("Distance", fontsize=22, fontweight='bold', fontname='Arial')
    plt.ylabel("Nematic Correlation", fontsize=22, fontweight='bold', fontname='Arial')
    plt.xticks(ticks=[0, 10, 20, 30], fontsize=18)
    plt.yticks(ticks=[0.0, 0.25, 0.5, 0.75, 1], fontsize=18)
    plt.legend(loc='upper right', fontsize=18, frameon=False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(1.25)
    plt.gca().spines['bottom'].set_linewidth(1.25)
    plt.tight_layout()
    plt.savefig(os.path.join(figures_folder, f"average_strain_{strain}_all_SCs_overlay-with-simus.png"), dpi=300, transparent=True)
    plt.savefig(os.path.join(figures_folder, f"average_strain_{strain}_all_SCs_overlay-with-simus.svg"), dpi=300, transparent=True)
    plt.close()

# Define plotting order (e.g., by strain ID)
#plot_order = ["1047", "1695", "459", "232", "177"]
#plot_order = ["459", "232", "177"]
#plot_order = ["232", "177"]
plot_order = ["436", "337"]

# Define legend order (custom order for display)
#legend_order = ["232", "177", "459", "1047", "1695"]
#legend_order = ["232", "177", "459"]
legend_order = ["337", "436"]

# Step 4: Plot average curves for each surface coverage, across all strains (based on selected strains)
for sc in sc_options:
    plt.figure(figsize=(4, 4))

    # Plot strains in the desired plot order
    for strain in plot_order:
        if strain not in strains_to_plot:
            continue
        key = (strain, sc)
        if key in data_by_strain_sc:
            dfs = data_by_strain_sc[key]
            combined_data = pd.concat([df for df, _ in dfs])
            avg_curve = combined_data.groupby("radius")["nematic_correlation"].mean()
            #plt.plot(
                #avg_curve.index,
                #avg_curve.values,
                #color=strain_colors[strain],
                #linewidth=4,
                #label=strain  # Temporary label
            #)

            # Plot theory curve if available
            theory_key = (strain, sc)
            if theory_key in theory_data_by_strain_sc:
                theory_df = theory_data_by_strain_sc[theory_key]
                plt.plot(
                    theory_df["radius"],
                    theory_df["nematic_correlation"],
                    linestyle='--',
                    linewidth=4,
                    color=strain_colors[strain],
                    label=f"Theory - Strain {strain}"
                )


    legend_elements = []
    for strain in legend_order:
        if strain in strains_to_plot:
            legend_elements.append(
                Line2D(
                    [0], [0],
                    color=strain_colors[strain],
                    lw=4, #strain_linewidths.get(strain, 6),  # or a fixed linewidth
                    linestyle='--',  # force plain line in legend
                    label=strain_labels[strain] #replace_legend_label(strain)
                )
            )

    #plt.legend(handles=legend_elements, loc='upper right', fontsize=16, frameon=False, bbox_to_anchor=(1.2, 1.2)) #bbox_to_anchor=(1.05, 1.05))




    # Plot formatting
    plt.xlim(0, 35)
    plt.ylim(0, 1)
    plt.xlabel('') #"Distance", fontsize=22, fontweight='bold', fontname='Arial')
    plt.ylabel('') #"Nematic Correlation", fontsize=22, fontweight='bold', fontname='Arial')
    plt.xticks(ticks=[0, 10, 20, 30], fontsize=22)
    plt.yticks(ticks=[0.0, 0.25, 0.5, 0.75, 1], fontsize=22)
    plt.tick_params(axis='x', labelsize=22)
    plt.tick_params(axis='y', labelsize=22)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(1.25)
    plt.gca().spines['bottom'].set_linewidth(1.25)
    plt.gca().tick_params(axis='both', which='minor', length=5, width=1.25, color='black')
    plt.gca().tick_params(axis='both', which='major', length=12, width=1.25, color='black')
    plt.gca().set_box_aspect(1)
    plt.tight_layout()
    plt.savefig(os.path.join(figures_folder, f"average_all_strains_{sc}_overlay-with-simus.png"), transparent=True, dpi=300)
    plt.savefig(os.path.join(figures_folder, f"average_all_strains_{sc}_overlay-with-simus.svg"), transparent=True)
    #plt.close()

```
